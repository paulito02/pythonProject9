```python
df1 = pd.DataFrame(df,columns=["Name", "Wage", "Value"])
def Value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 10000000000
    return 0.0
wage = df['Wage'].replace('[\€]', '', regex=True).apply(Value_to_float)
value = df['Value'].replace('[\€]', '', regex=True).apply(Value_to_float)
df1['Wage'] = wage
df1['Value'] = value
df1["difference"] = df1["Value"] - df1["Wage"]
df1.sort_values('difference', ascending=False)
```

```python
# using bokehit gives more visualisation
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLSTIPS = HoverTool(tooltips=[
                    ("index", "$index"),
                    ("(Wage,Value)", "($Wage, $Value)"),
                     ("Name", "@Name"),]
                     )


p = figure(title="soccer 2019". x_axis_label='Wage', y_axis_label='Value', plot_width=700. plot_width=700, tools=(TOOLTIPS))
p.circle('Wage', 'Value', size=10, source=df1)
show(p)
```
